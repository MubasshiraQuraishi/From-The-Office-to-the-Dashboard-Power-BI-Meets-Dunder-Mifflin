```python
import datetime
date_columns = ['returned_at','purchased_at','cancelled_at','reviewed_at']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='%Y-%m-%d %H:%M:%S', errors = 'coerce')

df['returned_date'] = df['returned_at'].dt.date
df['purchased_date'] = df['purchased_at'].dt.date
df['cancelled_date'] = df['cancelled_at'].dt.date
df['reviewed_date'] = df['reviewed_at'].dt.date
```
